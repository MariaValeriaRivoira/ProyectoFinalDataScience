POSIBLES TECNICAS DE LIMPIEZA:

1. ELIMINACION DE COLUMNAS

Esta técnica es sencilla. Eliminamos toda la columna o característica con datos faltantes, lo que sin duda provocará una pérdida de información. Por lo que sólo debemos realizar esto cuando estemos seguros de que los datos que faltan no son informativos. De lo contrario, deberíamos considerar otras soluciones.



En este data set cuantas columnas con datos faltantes tenemos del total?
De esas columnas con datos faltantes cual es el porcentaje en cada una?

```python
# Calcula la cantidad de valores nulos por columna
num_missing = df.isna().sum()

# Filtra las columnas con valores faltantes
columns_with_missing = num_missing[num_missing > 0]

# Número total de columnas con datos faltantes
total_columns_with_missing = len(columns_with_missing)

# Calcula el porcentaje de valores faltantes en cada columna
total_rows = df.shape[0]  # Número total de filas en el DataFrame
percentage_missing = (num_missing / total_rows) * 100

# Imprime el número total de columnas con datos faltantes
print("Número total de columnas con datos faltantes:", total_columns_with_missing)
print("\nPorcentaje de valores faltantes por columna:")
for column in columns_with_missing.index:
    print(f"{column}: {percentage_missing[column]:.2f}%")
```

 ¿Cuáles son las columnas a las que les faltan más del 30% de datos?

```python
# Filtra las columnas con más del 30% de valores faltantes
columns_with_high_missing = columns_with_missing[percentage_missing > 30]

# Número total de columnas con más del 30% de valores faltantes
total_columns_with_high_missing = len(columns_with_high_missing)

# Imprime el nombre de las columnas con más del 30% de valores faltantes
print("Columnas con más del 30% de valores faltantes:")
for column in columns_with_high_missing.index:
    print(column)
```

En este caso a las columnas Churn Category y Churn Reason les falta un 73,46% de datos. La columna Churn Category no tiene mucha importancia, pero la columna Churn Reason posee datos relevantes, por lo tanto se borra la columna Churn Category sola.

```python
# Elimina la columna Churn Category
df_cleaned=df.drop(['Churn Category'], axis=1)
df=df_cleaned
```

3. IMPUTACION DE VALORES

En lugar de descartar datos, también podemos reemplazar los que faltan. Un método sencillo es imputar los faltantes con valores constantes.Además de las constantes, también podemos imputar los valores faltantes con estadísticas.

Por ejemplo, podemos imputar las columnas numéricas con sus respectivas medianas.

```python
df_copy = df.copy()
numeric_cols = df_copy.select_dtypes(include=['number']).columns.tolist()
med = df_copy[numeric_cols].median()
df_copy[numeric_cols] = df_copy[numeric_cols].fillna(med)
df=df_copy
```

También podemos imputar las columnas no numéricas con sus valores más frecuentes.

```python
# Identificar las columnas no numéricas
non_numeric_cols = df.select_dtypes(exclude=['number']).columns.tolist()

# Calcular los valores más frecuentes en las columnas no numéricas
most_freq = df[non_numeric_cols].mode().iloc[0]

# Imputar los valores más frecuentes en las columnas no numéricas
df[non_numeric_cols] = df[non_numeric_cols].fillna(most_freq)
```

### Detección y manejo de valores innecesarios

**Tipo innecesario n.° 1: repetitivo y poco informativo**

Una columna puede tener muchas observaciones con el mismo valor. Cuando un porcentaje extremadamente alto de la columna tiene un valor repetitivo, debemos investigar si dicha columna proporciona información valiosa.

¿Cómo saberlo?
Podemos generar una lista de columnas con un alto porcentaje del mismo valor. Por ejemplo, especificamos a continuación mostrar columnas con más del 99,9% de filas con el mismo valor.

```python
num_rows = len(df)

for col in df.columns:
    cnts = df[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]

    if top_pct > 0.999:
        print('{0}: {1:.2f}%'.format(col, top_pct*100))
        print(cnts)
        print()
```

De hecho, no existe tal columna en nuestro conjunto de datos.

Grafico que muestra la cantidad de clientes por ciudad. (solo las 5 ciudades donde se dieron de baja mas clientes)

```python
# Calcular la proporción de clientes por ciudad
categoria_proporcion = df['City'].value_counts() / len(df) * 100

# Tomar solo las 5 ciudades con mayor porcentaje
top_5_ciudades = categoria_proporcion.head(5)

# Crear el gráfico circular con las 5 ciudades principales
plt.figure(figsize=(5, 4))
plt.pie(top_5_ciudades.values, labels=top_5_ciudades.index, autopct='%1.1f%%')
plt.title('Proporción de clientes dados de baja por las 5 ciudades principales')
plt.axis('equal')
plt.show()
```
